## ttest_ind

**Doing the same relation with help of T-test Independent**

Hypothesis Conditions :

**H₀:  (Null Hypothesis)**

Budget has no impact on IMDb rating.

**H₁:  (Alternate Hypothesis)**

Higher budget → higher rating.

Do HIGH budget movies have different RATINGS compared to LOW budget movies?


T-test actually compares same variable in 2 groups.


So ,we split the data into 2 groups High and Low.

```python
# Code

from scipy.stats import ttest_ind
```

```python
median_budget = movies_cleaned['budget'].median()
```

```python
high = movies_cleaned[movies_cleaned['budget'] >= median_budget ]['IMDb_rating']
low = movies_cleaned[movies_cleaned['budget'] < median_budget ]['IMDb_rating']
```

```python
ttest , p_value = ttest_ind(high , low)
```

```python
print("T-test :", ttest)
print("P_Value :", p_value)
```

```python
# Checking Hypothesis :

alpha = 0.05
if p_value  < alpha:
  print("Reject Null Hypothesis")
  print("Budget has no impact on IMDb rating.")
else:
  print("Failed Null Hypothesis")
  print("Budget has impact on IMDb rating.")
```

**Do male and female directors receive different IMDb ratings?**

```python
# merge 2 tables :

movies_directors = movies.merge(directors, left_on='director_id' , right_on='id' , how="left")
movies_directors.head()
```

## Visualize

```python
sns.boxplot(x='gender' , y='IMDb_rating' , data=movies_directors)
plt.title("IMDB rating by Director Gender  (Hypothesis 2)")
plt.show()
```

## Perform ttest :

```python
from scipy.stats import ttest_ind

male = movies_directors[movies_directors['gender'] == 'Male']['IMDb_rating']
female = movies_directors[movies_directors['gender'] == 'Female']['IMDb_rating']

ttest , p_value = ttest_ind(male , female)

print("Ttest :" , ttest)
print("P_Value :" , p_value)
```

## Checking Hypothesis :

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis (H₀)")
    print("There IS a statistically significant difference in IMDb ratings between male and female directors.")
else:
    print("Fail to Reject Null Hypothesis (H₀)")
    print("There is NO statistically significant difference in IMDb ratings between male and female directors.")
```
